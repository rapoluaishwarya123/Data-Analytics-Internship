```python
import pandas as pd

data = {
    "Date": ["2026-04-01","2026-04-02","2026-04-03","2026-04-04","2026-04-05",
             "2026-04-06","2026-04-07","2026-04-08","2026-04-09","2026-04-10"],
    "Product": ["Mobile","Laptop","Shoes","T-shirt","Watch",
                "Headphones","Bag","Jeans","Tablet","Sunglasses"],
    "Category": ["Electronics","Electronics","Fashion","Fashion","Accessories",
                 "Electronics","Accessories","Fashion","Electronics","Accessories"],
    "Quantity": [2,1,3,5,2,4,3,2,1,4],
    "Price": [15000,50000,2000,800,3000,1500,2500,2000,20000,1000]
}

df = pd.DataFrame(data)

# Create Revenue column
df["Revenue"] = df["Quantity"] * df["Price"]

df
```

```python
total_revenue = df["Revenue"].sum()
total_quantity = df["Quantity"].sum()
avg_revenue = df["Revenue"].mean()

print("Total Revenue:", total_revenue)
print("Total Quantity:", total_quantity)
print("Average Revenue:", avg_revenue)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df["Date"], df["Revenue"])
plt.title("Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.bar(df["Product"], df["Revenue"])
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
category_data = df.groupby("Category")["Revenue"].sum()
```

```python
plt.figure()
plt.pie(category_data, labels=category_data.index, autopct='%1.1f%%')
plt.title("Revenue by Category")
plt.show()
```

```python
df.to_csv("sales_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Line Chart
axs[0,0].plot(df["Date"], df["Revenue"])
axs[0,0].set_title("Revenue Trend")
axs[0,0].tick_params(axis='x', rotation=45)

# Bar Chart
axs[0,1].bar(df["Product"], df["Revenue"])
axs[0,1].set_title("Revenue by Product")
axs[0,1].tick_params(axis='x', rotation=45)

# Pie Chart
category_data = df.groupby("Category")["Revenue"].sum()
axs[1,0].pie(category_data, labels=category_data.index, autopct='%1.1f%%')
axs[1,0].set_title("Revenue by Category")

# Hide empty space
axs[1,1].axis('off')

plt.tight_layout()
plt.show()
```
